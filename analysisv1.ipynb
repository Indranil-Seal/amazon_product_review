```python
col_list = list(amadata.columns)
datastr = pd.DataFrame(columns = ['colname','obs','uniques','missing_','dtype'])
for col in col_list:
    datum = pd.DataFrame({'colname':[col],
                          'obs':[len(amadata[col])],
                          'uniques':[len(amadata[col].unique())],
                          'missing_':[((amadata[col].isna().sum())/amadata.shape[0])*100],
                          'dtype':[amadata[col].dtype]})
    datastr = pd.concat([datastr,datum],axis = 0)
```

```python
amadata['brand_cat_key'] = amadata['brand'].astype(str) +"-"+amadata['categories'].astype(str)
brand_cat_list = list(amadata['brand_cat_key'].unique())
```

```python
len(brand_cat_list)
```

```python
colorsdata = pd.DataFrame()
for brandcat in brand_cat_list:
    datum = pd.DataFrame({'brand_category':brandcat,
                          'colors': amadata.loc[amadata['brand_cat_key'] == brandcat].colors.unique(),
                          '#missing':amadata.loc[amadata['brand_cat_key'] == brandcat].colors.isna().sum()})
    colorsdata = pd.concat([colorsdata,datum],axis = 0)
```
